```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.countplot(data=tips,y='day',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True, color='0', marker='v')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True, palette='pastel')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(data=tips,x='day', y='total_bill', hue='sex', split=True, palette='pastel')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(data=tips,x='day', y='total_bill')
plt.show()
```

```python
#Con cat plot puedes tener cualquiera de los graficos anteriores, solo cambiando la variable kind= box, violin, swarm etc
#Es exactamente el mismo caso de displot, que tambien te facilita todos los graficos de distribucion cambiando la variable kind
plt.figure(figsize=(10,10))
sns.catplot(data=tips,x='day', y='total_bill', hue='sex', kind='box')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.catplot(data=tips,x='day', y='total_bill', hue='sex', kind='box', col='time')
plt.show()
```
