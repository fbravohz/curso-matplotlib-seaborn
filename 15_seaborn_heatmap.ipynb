```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr());
```

```python
sns.heatmap(tips.corr(), annot=True);
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm');
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5);
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='0');
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='0', vmin=0, vmax=1);
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='0', vmin=0, vmax=1, cbar=False);
```
