```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
#Se reciben los axes como un array en el que se accede por posiciones del index
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2,'g')
axes[1,0].plot(x,np.tanh(x)**2,'gray')
```

```python
#Recibiendo los axes en una tupla de tuplas
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2,'g')
ax5.plot(x,np.tanh(x)**2,'gray')
ax6.plot(x,np.sinc(x),'dimgray')
ax7.plot(x,np.cosh(x)**2,'orange')
ax8.plot(np.cosh(x),x,'violet')
#Mejora el espaciado entre los axes
fig.tight_layout()
```
