```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,10))

#Propiedades axes[0]
#utilizar $__$ nos permite introducir notacion matematica como x^2
axes[0].plot(x,y,label="$sin(x)^2$")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
#Propiedades axes[1]
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='lower left',bbox_to_anchor=(0.1,0.1))
```
