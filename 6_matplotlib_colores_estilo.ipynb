```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
fig, axes =plt.subplots(figsize=(8,8))
axes.plot(x,x+1,'r')
axes.plot(x,x+2,'b')
axes.plot(x,x+3,'y')
axes.plot(x,x+4,'g')
```

MODIFICACIONES TIPO: MATLAB

```python
fig, axes =plt.subplots(figsize=(8,8))
axes.plot(x,x+1,'r--')
axes.plot(x,x+2,'b-')
axes.plot(x,x+3,'y.-')
axes.plot(x,x+4,'go:')
```

MODIFICACIONES TIPO: PYPLOT

```python
fig, axes =plt.subplots(figsize=(8,8))
axes.plot(x,x+1,color='green',alpha=0.8, linewidth=5, linestyle='-',marker='o', markersize=10, markerfacecolor='white')
axes.plot(x,x+2,color='blue', linewidth=8, linestyle='--',marker='x', markersize=10, markerfacecolor='white')
axes.plot(x,x+3,color='#66FF66', linewidth=2, linestyle='dashed',marker='v', markersize=10, markerfacecolor='white')
axes.plot(x,x+4,color='#CCFFCC', linewidth=4, linestyle=':',marker='s', markersize=10, markerfacecolor='white')
```
